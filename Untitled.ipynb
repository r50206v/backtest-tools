```python
import pandas as pd
import numpy as np
import bt
%matplotlib inline
```

```python
class SelectWhere(bt.Algo):
    def __init__(self, signal, sign):
        self.signal = signal
        self.sign

    def __call__(self, target):
        if target.now in self.signal.index:
            sig = self.signal.ix[target.now]
            selected = list(sig.index[sig])
            target.temp['selected'] = selected
            if selected:
                weights = self.signal.ix[target.now].copy(deep=True)
                weights = signal
                target.temp = weights
        return True
```

```python
data = pd.DataFrame(
    np.random.normal(loc=0, scale=3, size=(100, 2)) +\
    np.concatenate([np.arange(100).reshape(-1, 1), np.arange(100).reshape(-1, 1)], axis=1), 
    index=pd.date_range(start='2018-01-01', periods=100, freq='D'),
    columns=['Taiwan', 'China']
)
```

```python
data.plot()
```

```python
sig = data.copy(deep=True)
sig.loc['2018-02-03':] = 1
sig.loc['2018-03-15':] = -1
sig.loc[:'2018-02-03'] = 0

hold = data.copy(deep=True)
hold.loc[:,:] = 0
hold_1 = hold.loc['2018-02-27':'2018-03-05']
hold_2 = hold.loc['2018-03-24':'2018-04-01']

sig.loc['2018-02-27':'2018-03-05'] = 0
sig.loc['2018-03-24':'2018-04-01'] = 0
sig.plot()
```

```python
s1 = bt.Strategy('s1', [
    bt.algos.RunDaily(),
    bt.algos.SelectAll(),
    bt.algos.WeighEqually(),
    bt.algos.Rebalance()
])
s = bt.Strategy('s', [
    bt.algos.RunDaily(),
    bt.algos.SelectAll(),
    bt.algos.WeighTarget(sig),
    bt.algos.Rebalance()
])
test1 = bt.Backtest(s1, data)
test = bt.Backtest(s, data)
res = bt.run(test)
```

```python
from bt.backtest import Backtest, Result
def RunBacktest(*backtests):
    '''
    adjusted bt.run
    '''
    for bkt in backtests:
        bkt.run()
    return Result(*backtests)
```

```python
composite = RunBacktest(*[test1, test])
composite.display()

data = bt.merge(*[v.prices for v in composite.values()])
s = bt.Strategy('composite', [bt.algos.SelectAll(),
                      bt.algos.WeighInvVol(),
                      bt.algos.Rebalance()])

# create and run
t = bt.Backtest(s, data)
res = bt.run(t)
```
